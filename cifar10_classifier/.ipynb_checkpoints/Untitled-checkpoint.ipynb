```python
import torch
import torchvision
import torchvision.transforms as transforms
import torch.optim as optim
import torch.nn as nn

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import confusion_matrix
import numpy as np

from PIL import Image
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

Normalization of pixel values of each RGB channel
(x - mean) / std -> (pixel - 0.5) / 0.5

.ToTensor() makes the pixel values in the range (0.0, 1.0), and after normalization they will be in the range (-1.0, 1.0)

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5),(0.5, 0.5, 0.5))
])
```

A bit like train_test_split, but CIFAR-10 is already split, so the assignment of training and test data is via train=True/False

The model will be trained on 32 images at once (batch_size)

Shuffle=False does not "mix" the data (data can be shuffled for training, but not for testing)

```python
train_set = torchvision.datasets.CIFAR10(root='./data', train=True, 
                                         download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_set, batch_size=32, shuffle=False)

test_set = torchvision.datasets.CIFAR10(root='./data', train=False, 
                                         download=True, transform=transform)
test_loader = torch.utils.data.DataLoader(test_set, batch_size=32, shuffle=False)

classes = train_set.classes
```

- Running the constructor of the base class nn.Module
- Adding a convolutional layer (the input has 3 RGB channels, the output has 16 channels, i.e. 16 filters, filter size 3x3, paddig means that the output image has the same size as the input, because a margin of 1px is added
- Adding a max pooling layer - takes the largest value from each 2x2 square, reduces the image size by half 32x32 -> 16x16 -> 8x8
- The second convolutional layer has 16 channels at the input, 32 channels at the output, i.e. deeper features, the rest is the same
- The last layer is 128 inputs and 10 outputs (because there are 10 classes in CIFAR-10)

```python
class SimpleCNN(nn.Module): 
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3,16,3, padding=1)
        self.pool = nn.MaxPool2d(2,2)
        self.conv2 = nn.Conv2d(16,32,3, padding=1)
        self.fc1 = nn.Linear(32*8*8, 128)
        self.fc2 = nn.Linear(128,10)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 32*8*8)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

Create an instance of the model (i.e. SimpleCNN class) and transfer the model to GPU (if available)

Criterion is a cost function, it tells how much the model is wrong

The optimizer is an algorithm that updates the model weights to reduce the loss

```python
model = SimpleCNN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

Training loop
- Training for 10 epochs (one pass through all images in the training set)
- In each epoch, data from train_loader is processed in batches.
- Images and labels are transferred to device
- Resetting gradients from the previous batch (so they don't add up and cause bad results)
- Running images through the model
- Calculating the loss between prediction and label
- Calculating gradients (how each weight affects the loss)
- Updating the weight in the direction that reduces the loss (by the optimizer)
- Adding the loss to a variable to then calculate the average
- Every 100 batches, the average of the last 100 steps is printed

```python
for epoch in range(10):
    running_loss = 0.0
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data[0].to(device), data[1].to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        if i % 100 == 99:
            print(f"[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 100:.3f}")
            running_loss = 0.0
```

- Iterating over images from the test set in batches
- Disabling gradient calculation (they are not needed during testing, and this saves memory, so it speeds up the process)
- Transferring the batch of data to the device
- Running the images through the model
- For each image, the class with the highest predicted value is selected
- Counting correct predictions
- Transerring data from GPU to CPU and changing it into numpy

```python
correct = 0
total = 0

all_preds = []
all_labels = []

with torch.no_grad():
    for data in test_loader:
        images, labels = data[0].to(device), data[1].to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs,1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

print('Accuracy - test set:', round(100 * correct / total, 2))
```

Confusion Matrix

```python
cm = confusion_matrix(all_labels, all_preds)

plt.figure(figsize=(7,5))
sns.heatmap(cm, annot=True, xticklabels=classes, yticklabels=classes, cmap='Reds', fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

Visualization of model predictions on test images

- Iterator to loop over batches
- Get the first batch from the iterator (32 images and their labels)
- Move images and labels to device

- Unnormalization
- Moving tensor from GPU to CPU and changing to Numpy arrays
- Changing axes order to matplotlib axes

```python
dataiter = iter(test_loader)
images, labels = next(dataiter)
images, labels = images.to(device), labels.to(device)

def imgshow(img):
    img = img/2 + 0.5
    npimg = img.cpu().numpy()
    plt.imshow(np.transpose(npimg, (1,2,0)))
    plt.axis('off')

plt.figure(figsize=(14,7))
for i in range(10):
    plt.subplot(2,5, i+1)
    imgshow(images[i])
    true = classes[labels[i]]
    pred = classes[predicted[i]]
    plt.title(f'Pred: {pred} \nTrue: {true}')
plt.suptitle('Model predictions on test images')
plt.show()
```

```python
wrong_preds = (predicted != labels).nonzero(as_tuple=True)[0]

to_show = min(10, len(wrong_preds))
plt.figure(figsize=(14,7))
for i in range(to_show):
    wrong_pred_i = wrong_preds[i]
    plt.subplot(2,5, i+1)
    imgshow(images[wrong_pred_i])
    true = classes[labels[wrong_pred_i]]
    pred = classes[predicted[wrong_pred_i]]
    plt.title(f'Pred: {pred} \nTrue {true}')
    plt.axis('off')

plt.suptitle('Wrong predictions on test images')
plt.show()
```

Class prediction for other images

```python
image_path = ['corgi.jpg', 'cat.jpg', 'frog.jpg', 'airplane.jpg']

transform = transforms.Compose([
    transforms.Resize((32,32)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,0.5,0.5),
                        (0.5,0.5,0.5))
])

for image in image_path:
    img = Image.open(image).convert('RGB')
    img_tensor = transform(img).unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(img_tensor)
        _, predicted = torch.max(output,1)

    print(f'{image} - Prediction: {classes[predicted.item()]}')
```
